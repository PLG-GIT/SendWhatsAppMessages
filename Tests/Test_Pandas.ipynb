```python
import pandas as pd
from tkinter import filedialog as fd
```

```python
file_path = fd.askopenfilename(
    title='Selecione o arquivo Excel com a lista de destinatarios',
    filetypes=[
        ('Arquivo Excel','.xls'),
        ('Arquivo Excel','.xlsx')]
    )
```
